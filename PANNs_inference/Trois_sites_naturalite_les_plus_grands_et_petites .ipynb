# Avoir les 3 sites de naturalit√© la plus grande et la plus petite

------

```python
import pandas as pd
import numpy as np
```

```python

df = pd.read_csv('naturalness_index_by_site.csv')
```

```python
df.head(2).T
```

```python
# Supposons que ta DataFrame s'appelle df
df[['Transect', 'Dynamique', 'Position']] = df['CODE_TRANS'].str.split('_', expand=True)
```

```python
df
```

```python
df_transect = pd.read_csv('transect_commune_ld.csv')

df_transect
```

```python
df_transect.dtypes
```

```python
df.dtypes
```

```python
# Maintenant, change le type de la colonne 'Transect' en int64
df['Transect'] = df['Transect'].astype('int64')
```

```python
df = pd.merge(df, df_transect, on='Transect', how='inner')
df
```

-----

## LAYER = 4 :



```python
df_layer_4 = df[df['LAYER'] == 4].copy()
```

### Rayon = 100 :

```python

df_layer_4_sorted_descending_100 = df_layer_4.sort_values(by='mean_naturalness_100', ascending=False)

# S√©lection des trois premi√®res lignes correspondantes
df_layer_4_selected_rows_descending_100 = df_layer_4_sorted_descending_100.iloc[:3]

df_layer_4_sorted_ascending_100 = df_layer_4.sort_values(by='mean_naturalness_100', ascending=True)

# S√©lection des trois premi√®res lignes correspondantes
df_layer_4_selected_rows_ascending_100 = df_layer_4_sorted_ascending_100.iloc[:3]

# Affichage des r√©sultats
print("Trois premi√®res lignes avec les plus grands mean_naturalness_100 :\n", df_layer_4_selected_rows_descending_100)
print("Trois premi√®res lignes avec les plus petits mean_naturalness_100 :\n", df_layer_4_selected_rows_ascending_100)
```

* Les sites avec la plus grande naturalit√© :
- **564** : Couflens Saub√©
- **265** : Couflens Pouill
----
* Les sites avec la plus faible naturalit√© :
