```python
import os
import pandas as pd

from google.colab import drive
drive.mount('/content/gdrive')
```

```python
lon_min, lon_max, lat_min, lat_max = 0, 3, 41, 44

# Chemin vers le dossier "Valeurs acoustiques"
dossier_valeurs_acoustiques = "/content/gdrive/MyDrive/Valeurs acoustiques"

# Parcourir les dossiers "2022" et "2023"
annees = ["2022", "2023"]

for annee in annees:
    chemin_annee = os.path.join(dossier_valeurs_acoustiques, annee)

    # Parcourir les dossiers feuilles
    for dossier_parent in os.listdir(chemin_annee):
        chemin_dossier_parent = os.path.join(chemin_annee, dossier_parent)

        dfs_acoustic = []
        dfs_vggish = []
        print(chemin_dossier_parent)

        for dossier_feuille in os.listdir(chemin_dossier_parent):
            chemin_dossier_feuille = os.path.join(chemin_dossier_parent, dossier_feuille)

            # Vérifier si le dossier feuille contient le fichier acoustic_indices.csv
            acoustic_file = os.path.join(chemin_dossier_feuille, "acoustic_indices.csv")
            vggish_file = os.path.join(chemin_dossier_feuille, "VGGish acoustic indices.csv")

            nom_doss = dossier_parent+dossier_feuille

            print(acoustic_file)
            print(vggish_file)

            # Lire les fichiers et ajouter les colonnes 'dynamique' et 'position'
            if os.path.exists(acoustic_file):
                df_acoustic = pd.read_csv(acoustic_file)
                df_acoustic['dynamique'] = 'Bleu' if 'Bleu' in dossier_feuille else 'Rouge'
                df_acoustic['position'] = 'L' if 'L' in dossier_feuille else 'E'
                dfs_acoustic.append(df_acoustic)

            if os.path.exists(vggish_file):
                df_vggish = pd.read_csv(vggish_file)
                df_vggish['dynamique'] = 'Bleu' if 'Bleu' in dossier_feuille else 'Rouge'
                df_vggish['position'] = 'L' if 'L' in dossier_feuille else 'E'
                dfs_vggish.append(df_vggish)



        # Fusionner tous les DataFrames pour les indices acoustiques
        final_df_acoustic = pd.concat(dfs_acoustic, ignore_index=True)

        # Fusionner tous les DataFrames pour les VGGish acoustic indices
        final_df_vggish = pd.concat(dfs_vggish, ignore_index=True)

        final_df_acoustic.to_csv(os.path.join(chemin_dossier_parent,dossier_parent+"L_E_B_R_acoustic_indices.csv"))
        final_df_vggish.to_csv(os.path.join(chemin_dossier_parent,dossier_parent+"L_E_B_R_vggish_indices.csv"))



```
